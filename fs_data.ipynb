```python
#불러오기 - 첫번째, 두번째, 세번째 행들을 multiple header로 설정 / 인덱스은 첫번째 열(날짜)로 설정
fs = pd.read_csv("C:/Code/eda/fs_data.csv", encoding = 'euc-kr', header = [0, 1, 2], index_col=0) 
 
#Multiple header 이름과 index 이름 지정
fs.columns.names = ['code', 'name', 'category' ]
fs.index.name = 'date'

#Table의 value가 e형태로 나오지 않게 해주는 코딩
pd.options.display.float_format = '{:.2f}'.format
```

```python
#최초 불러오기 모습
print(fs.info())
fs.head()
```

```python
fs.tail()
```

```python
fs = fs.drop(["2019-11-01"]) # Null 값이 들어가 있는 마지막 행 제거
fs_dropna = fs.dropna(axis = 1) # 자본총계 ~ 당기순이익 열벡터 속 정보 중에 하나라도 Null 값이 있으면 전체 column 삭제. 상의 필요.
```

```python
fs_T = fs_dropna.stack(level = ['category']).T #자본총계 등의 정보만 stack하여 column으로 내리고, table transpose
fs_T = fs_T.stack(level = 'date') # 행벡터로 표시된 '날짜'를 stack하여 column으로 내림
fs_T.drop(['매출액', '영업이익'], axis = 1, inplace = True) #필요없는 정보 (매출액, 영업이익) column 삭제
```
